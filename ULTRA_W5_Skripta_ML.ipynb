# Průřez umělou inteligencí a strojovým učením

## Co je umělá inteligence?
- *Věda, zabývající se schopností počítačů dělat věci, které vyžadují inteligenci, pokud mají být dělané lidmi* (Marvin Minsky, 1967)
- **Snaha tvořit inteligentní entity**
- Univerzální obor:
    - Hraní her (Go, Šachy, Poker)
    - Řízení dopravních prostředků
    - Dokazování a objevování
    - Psaní knih, básní, článků
    - Diagnostika nemocí
    - Mluvení a poslouchání

Stojí na principech formalizace racionálního uvažování:
- Aristoteles (Zákony racionálního myšlení)
- René Decartes (Síla racionalizace a pochopení světa)
- George Boole (Formální logika)
Později začala těžit z probabilistických metod a pokroku v teorii pravděpodobnosti 
- Gerloamo Cardamo (Možné výsledky událostí v hazardu)
- Thomas Bayes (Pravidla pro aktualizaci pravděpodobností jevů ve světle nových vodítek)

Skládá se z řady "běžných" oborů, které ji využívají:
- Algoritmická teorie her (Game theory)
    - Studium dělání rozhodnutí, která netriviálně ovlivňují užitek (a tak i rozhodnutí) ostatních
    - Přímo navazuje na Ekonomii - John Forbes Nash Jr.
        - Studium dělání rozhodnutí za účelem dosažení cílů
    - John von Neumann, Oskar Morgenstern - Teorie her a ekonomické chování (1944)
- Teorie užitku (Utility theory)
    - Formální model toho co je "cílem"
- Teorie rozhodnutí (Decision theory)
    - jak dělat "nejistá" rozhodnutí (ignorujeme ostatní)
- Operační výzkum
    - Dělání rozhodnutí tam, kde užitek není zjevný okamžitě, ale až po sérii rozhodnutí
- Control theory
    - Návrh systémů, které maximalizují danou funkci během času

Přispěly do ní také pokroky v medicíně a dalších vědách:
- Paul Broca - Studium pacientů s poškozením mozku
    - Objev neuronů a jejich propojení
    - Jak shluky jednoduchých buněk vedou k myšlenkám, činům a vědomí? 
- Introspekce, Behavioralismus, Kognitivní psychologie
- Kenneth Craik - Knowledge based agent
    1. Stimulace je reprezentovaná interním stavem
    2. Interní stav je upravován kognitivními procesy, což vede k novému stavu
    3. Nový stav se přeloží zpět do jiné akce


**Umělá inteligence je snaha tvořit inteligentní entity** (agenty).
- Agent je cokoliv, co pozoruje své okolí pomocí nějakých sensorů a reaguje se svým okolím pomocí aktuátorů
- Racionální agent vybírá takovou akci, od které očekává, že nejvíce zlepší jeho efektivitu

$a' = \text{argmax}_{a \in A} \ \mathbb{E}(\text{utility}(p, a))$
- $a$ je akce
- $a'$ je zvolená akce
- $A$ je množina všech akcí
- $\text{utility}$ je užitková funkce
- $p$ je vjem získaný pomocí senzorů (domělý stav okolí agenta)


Agenti se pohybují a operují v prostředích, která mohou mít různé vlastnosti:
- Zcela / částečně pozorovatelné
- Deterministické / Stochastické
- Epizodické / Sekvenční
- Statické / Semi-Dynamické / Dynamické
- Diskrétní / Spojité
- Monoagentní / Multiagentní
    - Kompetetivní / Kooperativní

- Markovovský rozhodovací proces
- Hry a teorie her
- Strojové učení
    - Kategorizace
    - Regrese
    - Rozhodovací stromy
    - Zpětnovazební učení

## Filosofické a Etické problémy umělé inteligence

- Allaisův paradox: (Lidi silně přitahuje výhra, která je jistá)
    - A. 80% šance na výhru 4000USD
    - B. 100% šance na výhru 3000USD

    - C. 20% šance na výhru 4000USD
    - D. 25% šance na výhru 3000USD

- Ellsbergův paradox: Urna obsahuje z 1/3 modré míčky a ze 2/3 kombinaci zelených a červených míčků (Pro lidi je pohodlnější riskovat se známou pravděpodobností, než s neznámou)
    - A. 100USD za modrý míček
    - B. 100USD za zelený míček

    - C. 100USD za modrý nebo zelený míček
    - D. 100USD za zelený nebo červený míček


- Slabá / Silná AI
- Obecná / Specializovaná AI
