```python
import sqlite3
```

```python
# Create connection, cursor, tables
with sqlite3.connect("files/movies.db") as connection:
    cursor = connection.cursor()

    # Enable foreign keys
    cursor.execute("PRAGMA foreign_keys = ON")

    # For repeatability, drop tables if they exist
    cursor.execute("DROP TABLE IF EXISTS t_Movie")
    cursor.execute("DROP TABLE IF EXISTS t_Director")

    # Create Director table
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS t_Director (
            Id INTEGER PRIMARY KEY AUTOINCREMENT,
            vch_FirstName TEXT NOT NULL,
            vch_LastName TEXT NOT NULL,
            d_Birth DATE NOT NULL
        )
    """)

    # Create Movie table
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS t_Movie (
            Id INTEGER PRIMARY KEY AUTOINCREMENT,
            vch_Title TEXT NOT NULL,
            d_Premiere DATE,
            fk_DirectorId INTEGER,
                   
            FOREIGN KEY (fk_DirectorId) REFERENCES t_Director(Id)
        )
    """)

```
